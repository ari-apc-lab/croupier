```python
with open('metrics_mahti_last_month.csv', 'r') as f:
    lines = f.readlines()
    for line in lines[1:]:
        line_split = line.split(',')
        filename = line_split[0]
        index = 0
        for split in line_split[1:]:
            if '[' not in split or ']' not in split:
                filename = filename + '_' + line_split[index] + '_' + line_split[index+1]
                print(filename)
                break
            index += 1
            
        
```
